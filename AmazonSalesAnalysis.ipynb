```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
print("Libraries imported successfully.")
```

```python
df = pd.read_csv("cleaned_dataset.csv")
df
```

```python
print("\nDescriptive Statistics for Dataset:")
print(df.describe(include = "all"))
```

```python
print(f"Dataset shape:{df.shape[0]} rows and {df.shape[1]} columns.")
```

```python
print("/nMissing values before data cleaning.")
print(df.isnull().sum())
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
df.duplicated().sum()
d = df.drop_duplicates(inplace = True)
print(d)
```

```python
#Convert Date Column
df["order_date"] = pd.to_datetime(df["order_date"])
df
```

```python
#Extract Month & Year
df["month"] = df["order_date"].dt.month
df["year"] = df["order_date"].dt.year
df
```

```python
#Total Revenue
total_revenue = df["total_revenue"].sum()
print(total_revenue)
```

```python
#Orders that generate the most profit
top_profit_orders = df.sort_values(by="profit", ascending=False).head(10)

print(top_profit_orders[["order_id","product_category","profit"]])
```

```python
#Top Selling Product Categories
df.groupby("product_category")["quantity_sold"].sum().sort_values(ascending=False)
```

```python
#Region with Highest Sales
df.groupby("customer_region")["total_revenue"].sum().sort_values(ascending=False)
```

```python
#Top Profitable Products
df.groupby("product_category")["profit"].sum().sort_values(ascending=False).head(10)

```

```python
#Most Used Payment Method
df["payment_method"].value_counts()
```

```python
#Payment Methods(Pie Chart)

payment = df["payment_method"].value_counts()

payment.plot(kind="pie", autopct='%1.1f%%')

plt.title("Payment Method Distribution")

plt.show()

# Shows which payment customers prefer.
```

```python
#Monthly Sales Trend (Line Chart)

monthly_sales = df.groupby("month")["total_revenue"].sum()

monthly_sales.plot(kind="line",marker="o",color='mediumseagreen')

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.show()

# Helps identify seasonal sales trends
```

```python
# Monthly Profit Trend (Line Chart)

monthly_profit = df.groupby("month")["profit"].sum()

plt.figure(figsize=(10,6))

plt.plot(monthly_profit.index, monthly_profit.values, marker='o',color='mediumseagreen')

plt.title("Monthly Profit Trend")
plt.xlabel("Month")
plt.ylabel("Total Profit")

plt.show()

# Shows seasonal profit trends across months.
```

```python
#Product Price Distribution (Histogram)

plt.figure(figsize=(10, 6))

plt.hist(df["price"], bins=20, edgecolor='black',color='mediumseagreen')

# Add titles and labels
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.show()


# Shows which price ranges appear most frequently in the dataset.
```

```python
#Customer Rating Distribution

plt.figure(figsize=(10,6))

sns.histplot(df["rating"], bins=10, kde=True,color = 'mediumseagreen')

plt.title("Customer Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Frequency")

plt.show()

# kde=True adds a smooth curve to show rating trends.
# A Customer Rating Distribution shows how customer ratings are spread
```

```python
# Profit Distribution (Histogram)

plt.figure(figsize=(10, 6))

plt.hist(df["profit"], bins=20, color='mediumseagreen', edgecolor='black')

plt.title('Distribution of Profit')
plt.xlabel('Profit')
plt.ylabel('Frequency')

plt.show()

#Helps identify low profit vs high profit orders.
```

```python
# Discount vs Revenue (Scatter Plot)

plt.scatter(df["discount_percent"], df["total_revenue"],color='mediumseagreen',edgecolor = "black")

plt.title("Discount vs Revenue")
plt.xlabel("Discount %")
plt.ylabel("Revenue")

plt.show()

# Shows if higher discounts increase sales.
```

```python
# Revenue by Category (Box Plot)

sns.boxplot(x="product_category", y="total_revenue", data = df,color='mediumseagreen')

plt.xticks(rotation=90)

plt.show()


# Shows revenue spread for each category.
```

```python
# Discount Impact on Sales (Box Plot)

plt.figure(figsize=(10,6))

sns.boxplot(x=df["discount_percent"], y=df["total_revenue"],color='mediumseagreen')

plt.title("Impact of Discount on Revenue")
plt.xlabel("Discount Percent")
plt.ylabel("Revenue")

plt.show()

# Helps analyze how different discount levels affect revenue.
```

```python
# Rating vs Revenue (Scatter Plot)

plt.figure(figsize=(10,6))

plt.scatter(df["rating"], df["total_revenue"],color='mediumseagreen')

plt.title("Customer Rating vs Revenue")
plt.xlabel("Rating")
plt.ylabel("Total Revenue")

plt.show()

# Shows whether higher rated products generate more revenue.
```

```python
# Correlation Analysis (Heatmap)

corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.show()

# Shows relationships between variables
```

```python
#Top Product Categories(Bar Chart)

category_sales = df.groupby("product_category")["quantity_sold"].sum()

category_sales.plot(kind="bar", figsize=(10,6), color='mediumseagreen',edgecolor = "black")

plt.title("Units Sold per Product Category")
plt.xlabel("Product Category")
plt.ylabel("Units Sold")

plt.xticks(rotation=45)

plt.show()

```

```python
# Sales by Region (Bar Chart)

region_sales = df.groupby("customer_region")["total_revenue"].sum()

plt.figure(figsize=(10,6))
plt.bar(region_sales.index, region_sales.values,color='mediumseagreen',edgecolor = "black")

plt.title("Total Revenue by Region")
plt.xlabel("Customer Region")
plt.ylabel("Total Revenue")

plt.xticks(rotation=45)

plt.show()

# Shows which regions generate the most revenue, helping businesses focus marketing efforts
```

```python
# Profit by Category (Bar Chart)

category_profit = df.groupby("product_category")["profit"].sum()

plt.figure(figsize=(10,6))
plt.bar(category_profit.index, category_profit.values, color='mediumseagreen',edgecolor = "black")

plt.title("Total Profit by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Profit")

plt.xticks(rotation=45)

plt.show()

# Identifies most profitable product categories
```
